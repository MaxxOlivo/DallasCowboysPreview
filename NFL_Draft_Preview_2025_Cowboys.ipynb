# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as stats
```

# Import Dataset

```python
years = [2020, 2021, 2022, 2023, 2024]
Passing_Yards = [4161,4800,3736,4397,3862]
Passing_Attempts = [639,647,556,614,637]
Rushing_Yards = [1788,2119,2298,1920,1705]
Rushing_Attempts = [430,473,531,468,429]


# Create a list of dictionaries, with each dictionary representing data for a specific year.
# Replace the values with your actual data.
Dallas_data = [
    {'Year': years[i], 'Team': 'Dallas', 'Passing_yards': Passing_Yards[i], 'Passing_attempts': Passing_Attempts[i], 'Rushing_yards': Rushing_Yards[i], 'Rushing_attempts':Rushing_Attempts[i]}
    for i in range(len(years))
]

# Convert the list of dictionaries to a Pandas DataFrame
Dallas_df = pd.DataFrame(Dallas_data)

Dallas_df
```

```python
# prompt: Using dataframe Dallas_df: lineplot

import plotly.express as px

# Create the line plot using Plotly Express
fig = px.line(Dallas_df, x='Year', y='Total_yards', title='Total Yards from 2020 to 2024 for Dallas Cowboys Offense')

# Customize the plot (optional)
fig.update_xaxes(title_text='Year')
fig.update_yaxes(title_text='Total Yards')

# Show the plot
fig.show()
```

```python
sns.histplot(data=Dallas_df,x='Passing_yards',bins=4,kde=True)
plt.title('Distribution of Passing Yards by Frequency')
plt.xlabel('Passing Yards')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(data=Dallas_df,x='Passing_yards',color='grey')
plt.title('Distribution of Passing Yards by Frequency')
plt.xlabel('Passing Yards')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.lineplot(data=Dallas_df,x='Year',y='Rushing_yards', color = 'green')
plt.title('Distribution of Rushing Yards by Year')
plt.xlabel('Year')
plt.ylabel('Rushing Yards')
plt.show()
```

```python
sns.lineplot(data=Dallas_df,x='Year',y='Passing_yards', color = 'orange')
plt.title('Distribution of Passing Yards by Year')
plt.xlabel('Year')
plt.ylabel('Passing Yards')
plt.show()
```

```python
sns.lineplot(data=Dallas_df,x='Year',y='Total_yards', color = 'red')
plt.title('Distribution of Total Yards by Year')
plt.xlabel('Year')
plt.ylabel('Total Yards')
plt.show()
```

```python
sns.scatterplot(data=Dallas_df,x='Passing_yards',y='Rushing_yards', hue='Year', palette='viridis')
plt.title('Scatterplot of Passing Yards vs Rushing Yards by Year')
plt.xlabel('Passing Yards')
plt.ylabel('Rushing Yards')
plt.show()
```

```python
sns.scatterplot(data=Dallas_df,x='Passing_attempts',y='Rushing_attempts', hue='Year', palette='viridis')
plt.title('Scatterplot of Passing Attempts vs Rushing Attempts by Year')
plt.xlabel('Passing Attempts')
plt.ylabel('Rushing Attempts')
plt.show()
```

```python
year_2020 = Dallas_df[Dallas_df['Year'] == 2020]
plt.figure(figsize=(6, 6))
plt.pie(
    [year_2020['Passing_attempts'].values[0], year_2020['Rushing_attempts'].values[0]],
    labels=['Passing attempts', 'Rushing attempts'],
    autopct='%1.1f%%',
    colors=['orange', 'green'],
    startangle=90,
)
plt.title('Passing Yards vs Rushing Yards in 2020')
plt.show()

# If you want to create a pie chart for each year, you can iterate through the years:
for year in Dallas_df['Year'].unique():
    year_data = Dallas_df[Dallas_df['Year'] == year]
    plt.figure(figsize=(6, 6))
    plt.pie(
        [year_data['Passing_attempts'].values[0], year_data['Rushing_attempts'].values[0]],
        labels=['Passing attempts', 'Rushing attempts'],
        autopct='%1.1f%%',
        colors=['orange', 'green'],
        startangle=90,
    )
    plt.title(f'Offensive Distribution in {year}')
    plt.show()
```
